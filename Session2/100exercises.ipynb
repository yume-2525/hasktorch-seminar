61. Consider 2 sets of points P0,P1 describing lines (2d) and a point p, how to compute distance from p to each line i (P0[i],P1[i])? (★★★)

```haskell
let p0 = asTensor ([[0,0],[1,1]] :: [[Float]])
let p1 = asTensor ([[0,1],[1,0]] :: [[Float]])
let p = asTensor([1,2] :: [Float])
let v = p1-p0
let w = p - p0
let vx = select 1 0 v
let vy = select 1 1 v
let wx = select 1 0 w
let wy = select 1 1 w
let a = Torch.abs (vx * wy - vy * wx)
let b = Torch.sqrt(vx*vx + vy*vy)
a / b
```

62. Consider 2 sets of points P0,P1 describing lines (2d) and a set of points P, how to compute distance from each point j (P[j]) to each line i (P0[i],P1[i])? (★★★)

```haskell
let p0 = asTensor ([[0,0],[1,1]] :: [[Float]])
let p1 = asTensor ([[0,1],[1,0]] :: [[Float]])
let p = asTensor([[1,2],[2,3],[(-1),0]] :: [[Float]])
let v = p1-p0
let w = (unsqueeze (Dim 1) p) - (unsqueeze (Dim0) p0)
let vx = unsqueeze (Dim 0) (select 1 0 v)
let vy = unsqueeze (Dim 0) (select 1 1 v)
let wx = select 1 0 w
let wy = select 1 1 w
let a = Torch.abs (vx * wy - vy * wx)
let b = Torch.sqrt(vx*vx + vy*vy)
a / b
```
